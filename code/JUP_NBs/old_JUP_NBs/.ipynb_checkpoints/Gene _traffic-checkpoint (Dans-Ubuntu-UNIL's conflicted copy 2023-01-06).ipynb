#### Get Gw X genes

```python
def Get_chrom_gene_info(gff_path, chroms, feature, key):
    
    """
    Get gene IDs from a GFF for a chromosome of interest
    
    - gff_path:    The path to the gff file
    - chroms:      List of chromosomes for which to extract genes
    - feature:     The type of feature to extract (CDS, mRNA, exon, gene, etc.)
    - key:         The key for the ID of the gene (e.g. ID, Parent, etc.) 
    
    """
    
    Gene_dict = {}
    
    with open(gff_path) as gff:
        
        for line in gff:
            chrom = line.split()[0]
            
            if chrom in chroms:   

                if line.split()[2] == feature:
                    ID_field = line.split()[8]
                    start = line.split()[3]
                    stop = line.split()[4]
                    
                    for info in ID_field.split(";"):
                        if info.split("=")[0] == key:
                            gene = info.split("=")[1]
            
                    #print(chrom, feature, gene)
                    Gene_dict[gene] = {}
                    Gene_dict[gene]["CHROM"] = chrom
                    Gene_dict[gene]["START"] = start
                    Gene_dict[gene]["STOP"] = stop
                
    return Gene_dict
    
```

***Genes retained on both X and Y***

The criterion to identify these genes are:

1. Present on the X and Y of Gw as single copy orthologs
2. On the sex chromosome ancestrally 

The second criterion should not do much here, as it is difficult to see how a gene would move to the sex chromosome after it was sex linked, and be on both gametologs. However it might reduce the number of genes we find just because they were not annotated properly in one or more of the outgroups. Let see how this criterion affects results. 


```python
def Orthofiner_parser(N0_path):
    
    """
    This function parses the "N0.tsv" file in the "Phylogenetic_Hierarchical_Orthogroups" results directory
    
    This file contains all orthogroups identified in the analysis and lists the names of the orthologous/paralogous 
    genes in each species. So I can use it to make a dictionary of orthogroups, which I can then use to find groups that 
    I am interested in. 
    
    Note - it assumes 6 taxa were used. 
    
    """
    
    DB_dict = {}
    
    with open(N0_path) as N0:
        
        header = next(N0).strip()
        Tax1 = header.split("\t")[3]  # Af
        Tax2 = header.split("\t")[4]  # Aq
        Tax3 = header.split("\t")[5]  # Ga
        Tax4 = header.split("\t")[6]  # GaY
        Tax5 = header.split("\t")[7]  # Gw
        Tax6 = header.split("\t")[8]  # GwY
        
        for line in N0:
            line = line.strip("\n")
            HOG = line.split()[0]
            DB_dict[HOG] = {}
            
            DB_dict[HOG][Tax1] = [i.strip() for i in line.split("\t")[3].split(",")]
            DB_dict[HOG][Tax2] = [i.strip() for i in line.split("\t")[4].split(",")]
            DB_dict[HOG][Tax3] = [i.strip() for i in line.split("\t")[5].split(",")]
            DB_dict[HOG][Tax4] = [i.strip() for i in line.split("\t")[6].split(",")]
            DB_dict[HOG][Tax5] = [i.strip() for i in line.split("\t")[7].split(",")]
            DB_dict[HOG][Tax6] = [i.strip() for i in line.split("\t")[8].split(",")]
            
    return DB_dict
    
    
    
```

```python
N0_path_PC = "/home/djeffrie/Dropbox/Sexy_sticklebacks/Y_chrom_comp/Data/Af_Aq_Ga_GaY_Gw_GwY/N0.tsv"
N0_path_laptop = "/Users/danieljeffries/Dropbox/Sexy_sticklebacks/Y_chrom_comp/Data/Af_Aq_Ga_GaY_Gw_GwY/N0.tsv"

N0_DB = Orthofiner_parser(N0_path_PC)
```

### Gene duplications on the X and Y

There are likely some genes with multiple copies on one or both sex chromosomes. There are several scenarios that could cause this. 

**For example:**

***If the X has one copy of a gene, but the Y has more than one:***
- that gene may have duplicated on the Y . . . 
- or there might have been an ancestral paralogous group, and one X copy has since been lost. 
- the same scenarios would fit if the X had multiple copies and the Y only one.

***If both sex chromosomes have the same number of copies of a particular gene (>1 copy)***
- Again this could be an ancestral paralogous group, 
- Or duplications on both the X and Y of an ancestrally single copy gene. 

The question is **how do we distinguish between these scenarios?** I guess we need to get at the information in the gene trees.

For now, lets see how many genes are in multiple-copies on one or both sex chromosomes.

```python
singleGwX_multiGwY = []
singleGwAuto_multiGwY = []

multiGw_singleGwY = []
multiGwX_singleGwY = []
multiGwAuto_singleGwY = []

multiGw_multiGwY = []
multiGwX_multiGwY = []
multiGwAuto_multiGwY = []

for HOG in N0_DB:
           
    if all(["" not in N0_DB[HOG]["Gw"], "" not in N0_DB[HOG]["GwY"]]):
        
        if all([len(N0_DB[HOG]["Gw"]) == 1,len(N0_DB[HOG]["GwY"]) > 1]):
            if N0_DB[HOG]["Gw"][0] in Gw_X_genes:
                singleGwX_multiGwY.append((N0_DB[HOG]["Gw"][0], N0_DB[HOG]["GwY"]))
            elif N0_DB[HOG]["Gw"][0] in Gw_auto_genes:
                singleGwAuto_multiGwY.append((N0_DB[HOG]["Gw"][0], N0_DB[HOG]["GwY"]))
                print(Gw_X_genes[N0_DB[HOG]["Gw"][0]]["CHROM"])
                
        ## if there are multiple X or autosomal genes, track where they are in the genome
        if all([len(N0_DB[HOG]["Gw"]) > 1,len(N0_DB[HOG]["GwY"]) == 1]):
            
            Gw_locations = []

           
            for gene in N0_DB[HOG]["Gw"]:
                
                if gene in Gw_X_genes:
                    Gw_locations.append(Gw_X_genes[gene]["CHROM"])
                elif gene in Gw_auto_genes:
                    Gw_locations.append(Gw_auto_genes[gene]["CHROM"])
                else:
                    print(gene, "not in dict")
                                       
                       
            X_copies = 0
            Auto_copies = 0
            for chrom in Gw_locations:
                if chrom == "chr12" or chrom == "chr19":
                    X_copies += 1
                else:
                    Auto_copies += 1
                    
            if Auto_copies == 0:
                multiGwX_singleGwY.append((N0_DB[HOG]["Gw"], N0_DB[HOG]["GwY"][0]))
            if X_copies == 0:
                multiGwAuto_singleGwY.append((N0_DB[HOG]["Gw"], N0_DB[HOG]["GwY"][0]))
            if Auto_copies >= 1 and X_copies >= 1:
                multiGw_singleGwY.append((N0_DB[HOG]["Gw"], N0_DB[HOG]["GwY"][0]))
                
                
        ### Finally, if there are multiple X/autosomal genes, AND multiple Y genes
        
        if all([len(N0_DB[HOG]["Gw"]) > 1,len(N0_DB[HOG]["GwY"]) > 1]):
            
            Gw_locations = []
           
            for gene in N0_DB[HOG]["Gw"]:
                
                if gene in Gw_X_genes:
                    Gw_locations.append(Gw_X_genes[gene]["CHROM"])
                elif gene in Gw_auto_genes:
                    Gw_locations.append(Gw_auto_genes[gene]["CHROM"])
                else:
                    print(gene, "not in dict")
                                       
            
            
            X_copies = 0
            Auto_copies = 0
            for chrom in Gw_locations:
                if chrom == "chr12" or chrom == "chr19":
                    X_copies += 1
                else:
                    Auto_copies += 1
                    
            if Auto_copies == 0:
                multiGwX_multiGwY.append((N0_DB[HOG]["Gw"], N0_DB[HOG]["GwY"]))
            if X_copies == 0:
                multiGwAuto_multiGwY.append((N0_DB[HOG]["Gw"], N0_DB[HOG]["GwY"]))
            if Auto_copies >= 1 and X_copies >= 1:
                multiGw_multiGwY.append((N0_DB[HOG]["Gw"], N0_DB[HOG]["GwY"]))
                
                
print("Found %s orthogroups with one X copy and multiple Y copies" % len(singleGwX_multiGwY))
print("Found %s orthogroups with one autosomal copy and multiple Y copies" % len(singleGwAuto_multiGwY))
print("Found %s orthogroups with multiple X copies and only 1 Y copy" % len(multiGwX_singleGwY))
print("Found %s orthogroups with multiple Autosomal copies and only 1 Y copy" % len(multiGwAuto_singleGwY))
print("Found %s orthogroups with multiple copies spread across both X and autosomes and only 1 Y copy" % len(multiGw_singleGwY))

print("Found %s orthogroups with multiple X copies and multiple Y copies" % len(multiGwX_multiGwY))
print("Found %s orthogroups with multiple Autosomal copies and multiple Y copies" % len(multiGwAuto_multiGwY))
print("Found %s orthogroups with multiple copies spread across both X and autosomes and multiple Y copies" % len(multiGw_multiGwY))
    
            
            
            
```

### Orphaned Y genes

How many genes on the Y do not have an identified ortholog in the Gw genome? 

```python
single_copy_Y_orhpans = []
multi_copy_Y_orhpans = []

for HOG in N0_DB:
     if all(["" in N0_DB[HOG]["Gw"], "" not in N0_DB[HOG]["GwY"]]):
            if len(N0_DB[HOG]["GwY"]) == 1:
                single_copy_Y_orhpans.append(N0_DB[HOG]["GwY"][0])
            elif len(N0_DB[HOG]["GwY"]) > 1:
                multi_copy_Y_orhpans.append(N0_DB[HOG]["GwY"])
                
print("Found %s single copy Y genes with no annotated homolog on the Gw X or autosomes" % len(single_copy_Y_orhpans))
print("Found %s multi-copy Y gene groups with no annotated homolog on the Gw X or autosomes" % len(multi_copy_Y_orhpans))
```

### Ok how many Y genes are accounted for in the above analysis? 

```python
Y_genes = 0

Y_genes_list = []

## Add the 1-to-1 orthologs
Y_genes_list += [i[1] for i in GwAutoGwY_one2ones]
Y_genes_list += [i[1] for i in GwXGwY_one2ones]

## Add the multi Gw/Gwx to single GwYs
Y_genes_list += [i[1] for i in multiGwX_singleGwY]
Y_genes_list += [i[1] for i in multiGwAuto_singleGwY]
Y_genes_list += [i[1] for i in multiGw_singleGwY]

## Add the multiple Y copy lists

for HOG in singleGwX_multiGwY:
    Y_genes_list += HOG[1]
    
## Add the multi Autos/X and multi Y list

for HOG in multiGwX_multiGwY:
    Y_genes_list += HOG[1]
    
for HOG in multiGwAuto_multiGwY:
    Y_genes_list += HOG[1]

## Add the orphaned genes, with no known orthologs

Y_genes_list += single_copy_Y_orhpans

for HOG in multi_copy_Y_orhpans:
    Y_genes_list += HOG

print("Accounted for %s genes out of %s genes in the OrthoFinder N0.tsv output file" % (len(Y_genes_list), len(Gw_Y_genes)))
```

### Gene duplications on the X and Y

```python
singleGaX_multiGaY = []
singleGaAuto_multiGaY = []

multiGa_singleGaY = []
multiGaX_singleGaY = []
multiGaAuto_singleGaY = []

multiGa_multiGaY = []
multiGaX_multiGaY = []
multiGaAuto_multiGaY = []

singleUnplacedGa_multiGaY = []
multiUnplacedGa_singleGaY = []
multiUnplacedGa_multiGaY = []

for HOG in N0_DB: ## for orthogroup
           
    if all(["" not in N0_DB[HOG]["Ga"], "" not in N0_DB[HOG]["GaY"]]):  # genes exist for Ga and GaY
        
        ## Single X / Autosomal copy and multiple Y copies
        
        if all([len(N0_DB[HOG]["Ga"]) == 1,len(N0_DB[HOG]["GaY"]) > 1]): # If N genes in Ga = 1 and N genes in GaY > 1
            if N0_DB[HOG]["Ga"][0] in Ga_X_genes: # if the Ga gene is on the X
                singleGaX_multiGaY.append((N0_DB[HOG]["Ga"][0], N0_DB[HOG]["GaY"])) # count as X - Y ortholog
            elif N0_DB[HOG]["Ga"][0] in Ga_auto_genes: # if it is on an autosome 
                singleGaAuto_multiGaY.append((N0_DB[HOG]["Ga"][0], N0_DB[HOG]["GaY"])) # count as auto - Y ortholog
            elif N0_DB[HOG]["Ga"][0] in unplaced_genes:
                singleUnplacedGa_multiGaY.append((N0_DB[HOG]["Ga"][0], N0_DB[HOG]["GaY"]))
            else:
                print(N0_DB[HOG]["Ga"][0], "not in dict")
                
        ## Multiple X or autosomal copies, and single Y copy
        
        if all([len(N0_DB[HOG]["Ga"]) > 1,len(N0_DB[HOG]["GaY"]) == 1]):
            
            Ga_locations = []

           
            for gene in N0_DB[HOG]["Ga"]:
                
                if gene in Ga_X_genes:
                    Ga_locations.append(Ga_X_genes[gene]["CHROM"])
                elif gene in Ga_auto_genes:
                    Ga_locations.append(Ga_auto_genes[gene]["CHROM"])
                elif gene in unplaced_genes:
                    unplaced_Ga_genes.append(gene[0])
                else:
                    print(gene, "not in dict (single/multi)")
                                       
                       
            X_copies = 0
            Auto_copies = 0
            unplaced_copies = 0
            for chrom in Ga_locations:
                if chrom == "NC_053230.1":
                    X_copies += 1
                elif chrom in autosomes:
                    Auto_copies += 1
                elif chrom in unplaced:
                    unplaced_copies += 1
                    
            if all([Auto_copies == 0, X_copies >=1, unplaced_copies == 0]): ## multiple X copies
                multiGaX_singleGaY.append((N0_DB[HOG]["Ga"], N0_DB[HOG]["GaY"][0]))
            elif all([Auto_copies >= 1, X_copies ==0]): ## 1 or more multiple autosomal copies (allows for additional unplaced copies)
                multiGaAuto_singleGaY.append((N0_DB[HOG]["Ga"], N0_DB[HOG]["GaY"][0]))
            elif all([Auto_copies >= 1, X_copies >=1]): ## 1 or more multiple X copies (allows for additional unplaced copies)
                multiGa_singleGaY.append((N0_DB[HOG]["Ga"], N0_DB[HOG]["GaY"][0]))
            elif all([Auto_copies ==0, X_copies ==0, unplaced_copies >= 1]): ## multiple unplaced copies only
                multiUnplacedGa_singleGaY.append((N0_DB[HOG]["Ga"], N0_DB[HOG]["GaY"][0]))
                
                
        ### Finally, if there are multiple X/autosomal genes, AND multiple Y genes
        
        if all([len(N0_DB[HOG]["Ga"]) > 1,len(N0_DB[HOG]["GaY"]) > 1]):
            
            Ga_locations = []
           
            for gene in N0_DB[HOG]["Ga"]:
                
                if gene in Ga_X_genes:
                    Ga_locations.append(Ga_X_genes[gene]["CHROM"])
                elif gene in Ga_auto_genes:
                    Ga_locations.append(Ga_auto_genes[gene]["CHROM"])
                elif gene in unplaced_genes:
                    unplaced_Ga_genes.append(gene[0])
                else:
                    print(gene, "not in dict (multi/multi)")
                                       
            
            
            X_copies = 0
            Auto_copies = 0
            unplaced_copies = 0
            for chrom in Ga_locations:
                if chrom == "NC_053230.1":
                    X_copies += 1
                elif chrom in autosomes:
                    Auto_copies += 1
                elif chrom in unplaced:
                    unplaced_copies += 1
                    
            if all([Auto_copies == 0, X_copies >=1, unplaced_copies == 0]): ## multiple X copies
                multiGaX_multiGaY.append((N0_DB[HOG]["Ga"], N0_DB[HOG]["GaY"]))
            elif all([Auto_copies >= 1, X_copies ==0]): ## 1 or more multiple autosomal copies (allows for additional unplaced copies)
                multiGaAuto_multiGaY.append((N0_DB[HOG]["Ga"], N0_DB[HOG]["GaY"]))
            elif all([Auto_copies >= 1, X_copies >=1]): ## 1 or more multiple X copies (allows for additional unplaced copies)
                multiGa_multiGaY.append((N0_DB[HOG]["Ga"], N0_DB[HOG]["GaY"]))
            elif all([Auto_copies ==0, X_copies ==0, unplaced_copies >= 1]): ## multiple unplaced copies only
                multiUnplacedGa_multiGaY.append((N0_DB[HOG]["Ga"], N0_DB[HOG]["GaY"]))
                
print("Found %s orthogroups with one X copy and multiple Y copies" % len(singleGaX_multiGaY))
print("Found %s orthogroups with one autosomal copy and multiple Y copies" % len(singleGaAuto_multiGaY))
print("Found %s orthogroups with multiple X copies (and possibly additional unplaced copies) and only 1 Y copy" % len(multiGaX_singleGaY))
print("Found %s orthogroups with multiple Autosomal copies (and possibly additional unplaced copies) and only 1 Y copy" % len(multiGaAuto_singleGaY))
print("Found %s orthogroups with multiple copies spread across both X and autosomes (and possibly additional unplaced copies) and only 1 Y copy" % len(multiGa_singleGaY))

print("Found %s orthogroups with multiple X copies (and possibly additional unplaced copies) and multiple Y copies" % len(multiGaX_multiGaY))
print("Found %s orthogroups with multiple Autosomal copies (and possibly additional unplaced copies) and multiple Y copies" % len(multiGaAuto_multiGaY))
print("Found %s orthogroups with multiple copies spread across both X and autosomes (and possibly additional unplaced copies) and multiple Y copies" % len(multiGa_multiGaY))

print("\nFound %s orthogroups with 1 Y copy, and multiple copies on unplaced scaffolds" % len(multiUnplacedGa_singleGaY))
print("\nFound %s orthogroups with multiple Y copies and a single copy on unplaced scaffolds" % len(singleUnplacedGa_multiGaY))
print("\nFound %s orthogroups multiple 1 Y copy, and multiple copies on unplaced scaffolds" % len(multiUnplacedGa_multiGaY))

    
            
            
            
```
