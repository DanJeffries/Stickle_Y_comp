## Has there been a secondary recomb event between X and Y AFTER recomb suppression? 

```python
from matplotlib import pyplot as plt
    
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams.update({'font.size': 14})
```

```python
Gw_dNdS_path = "/Users/dj20y461/Data_temp/Stickleback/Stickle_Y_comp/data/popgen_stats/Gw.dNdS.results"
```

### Chr19

```python
with open(Gw_dNdS_path) as Gw_dNdS:
    
    next(Gw_dNdS) # skip header
    
    chr19_pos = []
    chr19_dNdS = []
    
    for line in Gw_dNdS:
        
        if line.split()[1] == "chr19":
            pos = int(line.split()[2])
            dNdS = float(line.split()[6])
            
            if dNdS < 10:
                chr19_pos.append(pos)
                chr19_dNdS.append(dNdS)
                    
                              
```

```python
fig = plt.figure(figsize = (20,5))

plt.scatter(chr19_pos, chr19_dNdS, color = "black", s = 1)
plt.xlabel("Position")
plt.ylabel("dNdS")


plt.show()
```

### Chr12 dS

```python
with open(Gw_dNdS_path) as Gw_dNdS:
    
    next(Gw_dNdS) # skip header
    
    chr12_pos = []
    chr12_dS = []
    
    for line in Gw_dNdS:
        
        if line.split()[1] == "chr12":
            pos = int(line.split()[2])
            dS = float(line.split()[8])
            
            if dS > 0:
                chr12_pos.append(pos)
                chr12_dS.append(dS)
                    
                              
```

```python
fig = plt.figure(figsize = (20,5))

plt.scatter(chr12_pos, chr12_dS, color = "black", s = 1)
plt.ylim(-0.01,0.05)
plt.xlabel("Position")
plt.ylabel("dNdS")


plt.show()
```

```python
with open(Gw_fst_path) as Gw_fst:
    
    next(Gw_fst) # skip header
    
    chr12_fst_pos = []
    chr12_fst = []
    
    for line in Gw_fst:
        
        if line.split()[0] == "chr12":
            
            pos = int(line.split()[1])
            fst = float(line.split()[4])
            N_snps = int(line.split()[3])
            
            if N_snps >=3:
                chr12_fst_pos.append(pos)
                chr12_fst.append(fst)
                    
                              
```
